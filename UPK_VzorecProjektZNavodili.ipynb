```python
import pandas as pd
import pingouin as pg

# Preberite Excel datoteko
file_path = 'Podatki.xlsx'  # Zamenjajte z dejansko potjo do datoteke
df = pd.read_excel(file_path)

# Prikaz prvih nekaj vrstic za pregled
print(df.head())

# Filtriranje podatkov glede na vir
google_data = df[df['source'] == 'Google']
microsoft_data = df[df['source'] == 'Microsoft']

# PovpreÄja za Google
google_avg = google_data[['q1', 'q2', 'q3', 'q4']].mean()
print("\nPovpreÄne ocene za Google:")
print(google_avg)

# PovpreÄja za Microsoft
microsoft_avg = microsoft_data[['q1', 'q2', 'q3', 'q4']].mean()
print("\nPovpreÄne ocene za Microsoft:")
print(microsoft_avg)

# UteÅ¾i za dimenzije
utezi = {
    'q1': {'SC': 0.808},
    'q2': {'SC': 0.797},
    'q3': {'SC': 0.658},
    'q4': {'SC': 0.585},
}

# Funkcija za izraÄun SC
def izracun_sc(row, utezi):
    return sum(row[q] * utezi[q]['SC'] for q in ['q1', 'q2', 'q3', 'q4'])

# IzraÄun SC za vsakega testiranca
df['SC'] = df.apply(lambda row: izracun_sc(row, utezi), axis=1)

# Prikaz prvih nekaj vrstic z dodanim stolpcem SC
print(df.head())

# PovpreÄje SC za Google
google_avg_sc = df[df['source'] == 'Google']['SC'].mean()
print(f"\nPovpreÄna SC za Google: {google_avg_sc:.3f}")

# PovpreÄje SC za Microsoft
microsoft_avg_sc = df[df['source'] == 'Microsoft']['SC'].mean()
print(f"PovpreÄna SC za Microsoft: {microsoft_avg_sc:.3f}")

# Filtriranje podatkov glede na vir
google_data = df[df['source'] == 'Google']
microsoft_data = df[df['source'] == 'Microsoft']

# IzraÄun Cronbachove alfe za Google
cronbach_alpha_google = pg.cronbach_alpha(google_data[['q1', 'q2', 'q3', 'q4']])
print(f"Cronbachova alfa za Google: {cronbach_alpha_google[0]:.3f}")

# IzraÄun Cronbachove alfe za Microsoft
cronbach_alpha_microsoft = pg.cronbach_alpha(microsoft_data[['q1', 'q2', 'q3', 'q4']])
print(f"Cronbachova alfa za Microsoft: {cronbach_alpha_microsoft[0]:.3f}")

from scipy.stats import ttest_ind

# SC vrednosti za Google in Microsoft
google_sc_values = google_data['SC']
microsoft_sc_values = microsoft_data['SC']

# Izvedba t-testa
t_stat, p_value = ttest_ind(google_sc_values, microsoft_sc_values)

print(f"T-statistika: {t_stat:.3f}, P-vrednost: {p_value:.3f}")

if p_value < 0.05:
    print("Razlika med Google in Microsoft SC je statistiÄno znaÄilna (p < 0.05).")
else:
    print("Razlika med Google in Microsoft SC ni statistiÄno znaÄilna (p >= 0.05).")

```
