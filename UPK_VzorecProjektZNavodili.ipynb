## Povzetek <a class="anchor" id="povzetek"></a>

V tem projektu smo raziskovali in primerjali sintezo ter razpoznavo govora pri dveh vodilnih sistemih, ki ju ponujata Google in Microsoft. Osredotočili smo se na oceno sinteze govora z uporabo vprašalnika SUISQ, pri čemer smo analizirali dimenzijo Speech Characteristics, ki vključuje naravnost, tekočnost, zakasnitev in energičnost govora.

Razvili smo spletno aplikacijo, ki omogoča izvajanje testiranja in zbiranje podatkov. Podatke smo nato analizirali z uporabo statističnih metod, pri čemer smo ugotovili, da je Microsoftov sistem za sintezo govora prejel višjo povprečno uteženo oceno dimenzije SC (9.150) kot Googlov sistem (8.338), kar predstavlja približno 9,74 % boljšo oceno. Kljub temu razlika med sistemoma ni bila statistično značilna. Notranja skladnost vprašalnika, ocenjena s Cronbachovo alfo, je bila za oba sistema sprejemljiva, kar potrjuje zanesljivost uporabljenega instrumenta.

Projekt je omogočil poglobljen vpogled v delovanje in ocenjevanje tehnologij sinteze in razpoznavanja govora ter razkril priložnosti za prihodnje izboljšave, kot so razširitev vprašalnika z vprašanji za razpoznavo govora, večji vzorec testirancev in optimizacija postopka testiranja za boljšo uporabniško izkušnjo.

## 1. Uvod <a class="anchor" id="uvod"></a>


### 1.1. Opredelitev področja <a class="anchor" id="opredelitev-podrocja"></a>

Sinteza govora je postopek, pri katerem računalniški sistemi generirajo govor, ki posnema človeškega. Sisteme, ki omogočajo ta proces, imenujemo sintetizatorji govora (*speech synthesizers*), ti pa so na voljo bodisi kot programska bodisi kot strojna rešitev. Eden najpogostejših pristopov k sintezi govora je pretvorba besedila v govor (*text-to-speech*), kjer se napisano besedilo preoblikuje v naravni zvočni govor. Nasprotni proces pa predstavlja razpoznavanje govora, kjer sistem iz zvočnega posnetka prepozna govorjeno besedilo in ga pretvori v zapisano obliko.

Sinteza in razpoznavanje govora sta ključni tehnologiji, ki podpirata številne aplikacije na področjih, kot so pomoč uporabnikom, avtomatizacija procesov, in analiza podatkov. Uporaba teh tehnologij omogoča učinkovitejšo interakcijo med ljudmi in stroji, saj pretvarja glasovne ukaze v razumljive ukrepe (razpoznavanje govora) ali generira človeški govor na podlagi besedilnih vnosa (sinteza govora).

Orodij za sintezo in razpoznavo govora je več, v okviru projekta pa smo preizkusili delovanje orodij, ki ju ponujata Google in Microsoft. S tem smo pridobili vpogled v njihove zmogljivosti, učinkovitost ter odkrili morebitne omejitve pri uporabi.

### 1.2. Predstavitev problema <a class="anchor" id="predstavitev-problema"></a>

Razvoj naprednih rešitev za sintezo in razpoznavo govora zahteva premagovanje več izzivov. Sistemi morajo zanesljivo prepoznati govorne ukaze v različnih okoljih ali pri različni intonaciji govorca, generirani govor mora zveneti naravno in razumljivo, sistemi za razpoznavo in sintezo govora morajo biti razpoložljivi v različnih jezikih, kar je, kot bomo v videli v nadaljevanju, še vedno velik problem takih sistemov. Tudi če so te tehnične izzive uspešno premagali, ostaja ključna zahteva, da sistemi delujejo v realnem času, tudi na napravah z omejenimi računalniškimi viri.

## 2. Teoretično ozadje in trenutno stanje <a class="anchor" id="teoreticno-ozadje"></a>

Delovanje sinteze in razpoznave govora lahko razdelimo v več korakov. Pri sintezi govora prvi korak predstavlja pretvorba besedila. Sistem začne s pretvorbo vhodnega besedila v strukturo, ki je primerna za nadaljnjo obdelavo. Besedilo razčlenimo na manjše enote, kot so besede in stavki ali celo posamezni glasovi. Vhodno besedilo se nato normalizira oziroma pretvori v standardno obliko. Sistem nato določi meje med frazami ali stavki, kar omogoča bolj naravno zvočno izvedbo. Ta postopek se imenuje ločevanje intonacijskih skupin. 

V drugem koraku se osredotočamo na jezične lastnosti. Sistem analizira jezikovne značilnosti vhodnega besedila in določi, kako jih predstaviti v govoru. To obsega določanje naglasov, ki je pogosto dinamičen in odvisen glede oblike ali pomena besede, intonacijo, ki določa, kako se višina glasu spreminja med govorom, hitrost govora in melodija ter ritmika.

V zadnjem, tretjem koraku, se tvorijo zvočni signali, ki jih uporabnik sliši kot govor. Poznamo različne pristope sinteze govora, od katerih je odvisno, kako se končni signal sliši.

Pri sintezi z združevanjem gre za združevanje vnaprej posnetih segmentov govora. Segmenti so lahko posamezni fonemi, zlogi, besede ali fraze, ki jih sistem poveže glede na besedilo. Izhod takega sistema je naraven zvok, ki pa je zelo omejeno prilagodljiv, saj sistem ne more generirati novih zvokov izven baze.

Parametrična sinteza modelira glas s pomočjo parametrov, kot so višina tona, trajanje in amplituda, s katerimi generiramo zvok. Ponuja večjo prilagodljivost glasu, ampak je ta v primerjavi s sintezo z združevanjem manj naraven.

Tako najbolj prilagodljiv kot tudi naraven zvok pa nam omogočajo sistemi, ki uporabljajo globoko učenje, npr. Googlov WaveNet in Microsoftov Neural TTS. Ta uporablja nevronske mreže za tvorbo govora, pri čemer sistem predvideva naslednjo vrednost zvočnega vala na podlagi prejšnjih vrednosti. Naraven in prilagodljiv zvok pa nam v tem primeru otežuje visoka računska zahtevnost takih sistemov.

Razpoznavo govora bi lahko razdelili na štiri glavne korake. V prvem koraku najprej zajamemo zvok in ga pretvorimo v ustrezen format.

V drugem koraku, akustično modeliranje, zvočni signal analiziramo in prepoznamo osnovne enote jezika oziroma foneme. Akustični modeli povezujejo značilnosti zvočnega posnetka s fonemi, pri čemer največkrat uporabljajo prikrite Markovove modele ali globoke nevronske mreže. Akustično modeliranje omogoča razlikovanje med podobnimi zvoki in pravilno identifikacijo fonemov.

Tretji korak je jezikovno modeliranje. Na podlagi prepoznanih fonemov sistem uporablja jezikovne modele za sestavljanje besed in stavkov. Jezikovni modeli ocenjujejo verjetnosti določenih zaporedij besed, kar prispeva k zmanjšanju dvoumnosti in izboljšanju natančnosti razpoznavanja. Pogosto se uporabljajo statistični modeli, kot so n-grami, ali pa nevronske mreže in transformerji.

V zadnjem koraku sistem pretvori prepoznane besede in fraze v besedilo. Ta faza vključuje dodatno obdelavo za odpravljanje napak in zagotavljanje skladnosti besedila s pravopisnimi pravili.

Tudi pri razpoznavanju govora poznamo več različnih metod. Pri opazovanju ključnih besed sistem prepozna določene ključne besede v govoru. Uporablja se v aktivacijskih stavkih, kot je na primer "OK Google", kjer sistem prepozna le ključne fraze in ignorira vse ostalo.

Neprekinjeno prepoznavanje govora je proces razpoznavanja tekočega govora brez premorov med besedami, kar sistemu omogoča razpoznavo celotnih povedi oziroma daljšega govora.

Avtomatska razpoznava govora odvisna od govornika je sistem za razpoznavanje govora, ki je prilagojen specifičnemu govorniku. Takšni sistemi zahtevajo fazo treninga, kjer govorec zagotovi vzorčne podatke, da sistem prepozna značilnosti njegovega glasu. Avtomatska razpoznava govora neodvisna od govornika pa je sistem, ki lahko deluje z različnimi govorci brez potrebe po predhodnem prilagajanju in so oblikovani za razpoznavanje govora širokega spektra uporabnikov. Primeri takih sistemov so Siri, Alexa in Google Assistant.

Za naš projekt smo izbrali orodja sinteze in razpoznave govora Google in Microsoft, ki sta trenutno eni bolj razširjenih rešitev na področju sinteze in razpoznave govora.

Google Text-to-Speech API je storitev, ki temelji na tehnologiji WaveNet, razviti pri DeepMind. Uporabnik pošlje besedilo prek API-ja, kjer se to najprej razčleni na stavke, normalizira in obdeluje z analizo jezikovnih lastnosti, kot so naglas, intonacija, melodija in ritmika. Globoka nevronska mreža WaveNet nato generira zvočni signal, ki je naraven in prilagodljiv. Orodje omogoča tudi personalizacijo, vključno s prilagoditvami tona, hitrosti govora in glasnosti. Rezultat je govor v različnih formatih, kot sta MP3 ali LINEAR16.

Microsoft Azure orodje za sintezo govora sledi sorodnim korakom, vendar uporabljajo svojo tehnologijo Neural TTS. Ta prav tako temelji na globokih nevronskih mrežah in ponuja dodatno funkcionalnost personalizacije, uporabniki lahko tako na osnovi svojih posnetkov ustvarijo prilagojene glasove, kar omogoča večjo fleksibilnost pri uporabi.

Google Speech-to-Text API analizira zvočne posnetke, ki jih uporabnik naloži prek API-ja. Sistem uporablja kombinacijo konvolucijskih nevronskih mrež (CNN) za ekstrakcijo značilnosti zvočnih signalov in rekurentnih nevronskih mrež (RNN), kot so LSTM ali GRU. Zvočni posnetki se tako pretvorijo v fonetične prepise. Na tej stopnji se uporabljajo globoke nevronske mreže, kot so transformatorji, ki analizirajo kontekst govora, določajo pravilne besede in fraze ter odpravljajo dvoumnosti. Google omogoča prilagoditve za specifične jezike, dialekte ali ključne besede z uporabo funkcije Custom Speech Adaptation. Sistem vrne besedilo v realnem času ali kot zaključeno transkripcijo po obdelavi. Microsoftovo orodje deluje na podoben način, le da za akustično modeliranje uporablja svojo arhitekturo Deep Neural Networks, za jezikovno modeliranje pa uporablja Language Understanding Intelligent Service. Azure omogoča prilagoditev akustičnega in jezikovnega modela z uporabo funkcije Custom Speech, kjer lahko uporabniki trenirajo modele na svojih podatkih.

### 2.1. Teoretični vidiki problema <a class="anchor" id="teoreticni-vidiki-problema"></a>


Sistemi za sintezo in razpoznavo govora kot jih ponujata Google in Microsoft omogočajo visoko natančnost in naravnost, vendar se soočajo z že prej omenjenimi izzivi kot so podpora več jezikom, robustnost v hrupnih okoljih, resursna učinkovitost in splošna natančnost ter naravnost govora.

Podpora več jezikom
Večina sistemov ponuja veliko število jezikov za sintezo in razpoznavo govora, še vedno pa je podpora za manj razširjene jezike omejena, saj razvoj teh sistemov zahteva velike količine podatkov in procesiranja. Tako tudi Googlovo orodje za sintezo govora ne ponuja sintetizatorja v slovenskem jeziku. Microsoftovo orodje ima na voljo tudi sintezo v slonveskem jeziku, a je število glasov in ostalih nastavitev za slovenski jezik omejeno. Poleg tega predstavljajo problem dialekti in narečja, saj so tudi v podprtih jezikih regionalne različice jezika pogosto pomankljivo implementirane.

Robustnost v hrupnih okoljih
Uporabniki sisteme za razpoznavo govora pogosto uporabljajo v zunanjih, nepredvidljivih okoljih (npr. med vožnjo, v hrupnem okolju), sistemi za razpoznavo govora pa so občutljivi na okoljske motnje, kot so šum, več govornih virov hkrati in odmevi. Zato je potrebno razviti boljše algoritme za izolacijo glavnega zvočnega vira od ozadja, kar vključuje tehnike za izboljšanje signala, kot npr. beamforming ali noise-cancelation.

Resursna učinkovitost
Sistemi za sintezo in razpoznavo govora zahtevajo veliko računskih virov, kar omejuje njihovo uporabo na napravah, kot so pametni telefoni, IoT naprave in roboti. Del problema je rešen z uporabo API klicev, to pomeni, da se procesiranje izvaja na strani ponudnika storitev, medtem pa se končnemu uporabniku pošlje le rezultate. Kljub temu pa to od uporabnika zahteva nenehno povezavo z oblakom, zato so lahki in optimizirani modeli, ki lahko delujejo lokalno, brez povezave z oblakom, ključni. Primer orodja za sintezo govora, ki deluje lokalno je PiperTTS.

Splošna natančnost in naravnost govora
Končna natančnost prepoznave govora ostaja en glavnih izzivov sistemov za razpoznavo govora. Na to vplivajo zgoraj opisane stvari: dialekti, narečja, slabše podprti jeziki, hrupna okolja, itd. Zato so v bistvu prejšne točke se združijo v to točko - prispevajo k natančnosti in naravnosti govora. Pri razpoznavi govora imamo probleme kot je prilagajanje kontekstu, kar vpliva na interpretacijo pomena. Pri sintezi govora imamo izzive z intonacijo in prozodičnostjo; intonacija, melodija, poudarki in ritmičnost so včasih neustrezni.

Opisani so večji tehnični izzivi sinteze in razpoznave govora, omenimo pa lahko še izziv zasebnosti in etike. Sistemi, ki analizirajo ali shranjujejo govorne podatke, lahko predstavljajo tveganje za zasebnost uporabnikov. To še posebej velja za modele, kjer se procesiranje ne izvaja lokalno, ampak na strežnikih ponudnika storitve. Govorni modeli, ki temeljijo na nepopolnih podatkovnih zbirkah, lahko odražajo pristranskosti, povezane z jezikom, spolom, starostjo ali kulturo.

### 2.2. Algoritmi <a class="anchor" id="algoritmi"></a>

V sistemih za sintezo in razpoznavo govora se uporabljajo različni algoritmi in modeli, ki omogočajo učinkovito in natančno obdelavo govora. Med drugim se na tem področju pogosto uporabljajo prikriti markovovi modeli in nevronske mreže.

Prikriti Markovovi modeli (HMM)
Prikriti Markovovi modeli so statistični modeli, ki predstavljajo sisteme s skritimi stanji. V razpoznavi govora se uporabljajo za modeliranje zaporedij fonemov, kjer vsako stanje predstavlja določen fonem ali del fonema. HMM omogočajo verjetnostno modeliranje prehodov med stanji in opazovanji, kar je ključno za razumevanje govornega signala. 
[INFORMATICS EDINBURGH](https://www.inf.ed.ac.uk/teaching/courses/asr/2022-23/asr11-hybrid_hmm_nn.pdf?utm_source=chatgpt.com)

Globoke nevronske mreže (DNN)
Globoke nevronske mreže so sestavljene iz več plasti nevronov, ki omogočajo učenje kompleksnih vzorcev v podatkih. V akustičnem modeliranju se DNN uporabljajo za pretvorbo zvočnih značilnosti v verjetnosti fonemov, kar izboljšuje natančnost razpoznave govora. Kombinacija DNN in HMM, znana kot hibridni DNN-HMM pristop, združuje prednosti obeh metod za boljše rezultate. 
[SPRINGER LINK](https://link.springer.com/article/10.1007/s11704-020-9419-z?utm_source=chatgpt.com)

Konvolucijske nevronske mreže (CNN)
Konvolucijske nevronske mreže so posebej zasnovane za obdelavo podatkov z lokalno strukturo, kot so slike ali zvočni signali. V razpoznavi govora se CNN uporabljajo za ekstrakcijo prostorskih značilnosti iz spektrogramov, kar pomaga pri prepoznavanju fonetičnih vzorcev in izbolšanju robustnosti sistema v hrupnih okoljih.

Rekurentne nevronske mreže (RNN)
Rekurentne nevronske mreže so zasnovane za obdelavo zaporednih podatkov, saj imajo povezave, ki omogočajo shranjevanje informacij iz prejšnjih korakov. To je še posebej uporabno v razpoznavanju govora, kjer je kontekst zelo pomemben. Posebne oblike RNN, kot sta Long Short-Term Memory in Gated Recurrent Unit, se pogosto uporabljata za modeliranje dolgoročnih odvisnosti v govoru. 
[DE GRUYTER](https://www.degruyter.com/document/doi/10.1515/jisys-2018-0372/html?utm_source=chatgpt.com)

Transformatorji
Transformatorji so napredna arhitektura nevronskih mrež, ki temelji na mehanizmu pozornosti. Omogočajo učinkovito obdelavo dolgih odvisnosti v podatkih brez potrebe po zaporedni obdelavi, kar je značilno za RNN. V sintezi govora se transformatorji uporabljajo za generiranje naravnega in tekočega govora, saj lahko modelirajo kompleksne jezikovne strukture.

### 2.3. Merjenje uspešnosti rešitve <a class="anchor" id="merjenje-uspesnosti-resitve"></a>

Pri ocenjevanju učinkovitosti sistemov za sintezo in razpoznavo govora se uporabljajo različni pristopi in instrumenti, ki omogočajo kvantitativno in kvalitativno analizo njihovega delovanja. Za testiranje uporabniške izkušnje in celotnih sistemov se pogosto uporabljajo instrumenti, kot so SUS, UMUX, PANAS in SAM, vendar pa takšni instrumenti v našem primeru niso smiselni, saj naše testiranje meri zgolj pravilnost sinteze in razpoznavanja govora, ki v realnem sistemu predstavljata le del celotne uporabniške izkušnje.

Za merjenje lastnosti govornega vmesnika se pogosto uporablja SUISQ (Subjective Unnaturalness Interval Scale Questionnaire). SUISQ ocenjuje sistem na petih dimenzijah: usmerjenost k uporabnikovim ciljem (user goal orientation), vedenje do stranke (customer service behaviour), govorne značilnosti (speech characteristics), in zgovornost sistema (verbosity). Instrument vključuje 25 vprašanj na 5-stopenjski Likertovi lestvici, poleg tega pa obstajata tudi krajši različici vprašalnika s 14 in 8 vprašanji. Poleg SUISQ so na voljo tudi druge metode za ocenjevanje, ki se osredotočajo le na ocenjevanje uspešnosti sinteze in razpoznave govora.

Word error rate in Character error rate (WER in CER)
Word error rate meri delež napačno prepoznanih besed v primerjavi s celotnim številom besed v referenčnem besedilu. Izračuna se kot vsota vstavljanj, izbrisov in zamenjav, deljena s skupnim številom besed v referenčnem besedilu. Nižja vrednost WER pomeni boljšo natančnost sistema za razpoznavo govora. Character error rate je podoben WER, vendar meri napake na ravni znakov namesto besed in se zato uporablja pri sistemih, kjer je pomembna natančnost na ravni znakov.

Perceptual Evaluation of Speech Quality (PESQ)
Objektivna metoda za ocenjevanje kakovosti govora, ki primerja izvirni in sintetizirani ali prepoznani govor ter izračuna oceno kakovosti na podlagi modelov človeškega sluha.

Short-Time Objective Intelligibility (STOI)
Metrika, ki ocenjuje razumljivost govora z analizo kratkih časovnih segmentov signala. Pogosto se uporablja pri ocenjevanju učinkovitosti sistemov za izboljšanje govora ali pri sintezi govora v hrupnih okoljih.

ABX Test
Subjektivna metoda, kjer poslušalci primerjajo dva zvočna vzorca (A in B) ter določijo, kateri je bolj podoben referenčnemu vzorcu X. Uporablja se za ocenjevanje razlik med različnimi sistemi ali algoritmi za sintezo in razpoznavo govora.

Poleg tega poznamo še druge teste kot so Diagnostic rhyme test, Bilingual evaluation understudy score, Rapid prosody transcription, ASR-bootstraping, itd., pogosto pa se uporablja kombinacija več metod za boljšo oceno učinkovitosti teh sistemov.

```python
import pandas as pd
import pingouin as pg

# Preberite Excel datoteko
file_path = 'Podatki.xlsx'  # Zamenjajte z dejansko potjo do datoteke
df = pd.read_excel(file_path)

# Prikaz prvih nekaj vrstic za pregled
print(df.head())

# Filtriranje podatkov glede na vir
google_data = df[df['source'] == 'Google']
microsoft_data = df[df['source'] == 'Microsoft']

# Povprečja za Google
google_avg = google_data[['q1', 'q2', 'q3', 'q4']].mean()
print("\nPovprečne ocene za Google:")
print(google_avg)

# Povprečja za Microsoft
microsoft_avg = microsoft_data[['q1', 'q2', 'q3', 'q4']].mean()
print("\nPovprečne ocene za Microsoft:")
print(microsoft_avg)

# Uteži za dimenzije
utezi = {
    'q1': {'SC': 0.808},
    'q2': {'SC': 0.797},
    'q3': {'SC': 0.658},
    'q4': {'SC': 0.585},
}

# Funkcija za izračun SC
def izracun_sc(row, utezi):
    return sum(row[q] * utezi[q]['SC'] for q in ['q1', 'q2', 'q3', 'q4'])

# Izračun SC za vsakega testiranca
df['SC'] = df.apply(lambda row: izracun_sc(row, utezi), axis=1)

# Prikaz prvih nekaj vrstic z dodanim stolpcem SC
print(df.head())

# Povprečje SC za Google
google_avg_sc = df[df['source'] == 'Google']['SC'].mean()
print(f"\nPovprečna SC za Google: {google_avg_sc:.3f}")

# Povprečje SC za Microsoft
microsoft_avg_sc = df[df['source'] == 'Microsoft']['SC'].mean()
print(f"Povprečna SC za Microsoft: {microsoft_avg_sc:.3f}")

# Filtriranje podatkov glede na vir
google_data = df[df['source'] == 'Google']
microsoft_data = df[df['source'] == 'Microsoft']

# Izračun Cronbachove alfe za Google
cronbach_alpha_google = pg.cronbach_alpha(google_data[['q1', 'q2', 'q3', 'q4']])
print(f"Cronbachova alfa za Google: {cronbach_alpha_google[0]:.3f}")

# Izračun Cronbachove alfe za Microsoft
cronbach_alpha_microsoft = pg.cronbach_alpha(microsoft_data[['q1', 'q2', 'q3', 'q4']])
print(f"Cronbachova alfa za Microsoft: {cronbach_alpha_microsoft[0]:.3f}")

from scipy.stats import ttest_ind

# SC vrednosti za Google in Microsoft
google_sc_values = google_data['SC']
microsoft_sc_values = microsoft_data['SC']

# Izvedba t-testa
t_stat, p_value = ttest_ind(google_sc_values, microsoft_sc_values)

print(f"T-statistika: {t_stat:.3f}, P-vrednost: {p_value:.3f}")

if p_value < 0.05:
    print("Razlika med Google in Microsoft SC je statistično značilna (p < 0.05).")
else:
    print("Razlika med Google in Microsoft SC ni statistično značilna (p >= 0.05).")

```
